### <a id='reviewrating'>Number of Reviews and Average Rating</a>

Find the number of user reviews and the average rating for each book.

```python
# This query shows the first 15 rows. If you need the full table, please comment out (--) its last line (LIMIT 15;)

query = '''
SELECT
    DISTINCT reviews.book_id AS book_id,
    books.title,
    authors.author,
    COUNT(reviews.review_id) AS n_reviews,
    subquery.avg_rating,
    subquery.n_ratings
FROM reviews
LEFT JOIN
    (
    SELECT 
        DISTINCT book_id,
        AVG(rating) AS avg_rating,
        COUNT(DISTINCT rating_id) AS n_ratings
    FROM ratings 
    GROUP BY book_id
    ) AS subquery ON subquery.book_id = reviews.book_id
LEFT JOIN books ON books.book_id = reviews.book_id
LEFT JOIN authors ON authors.author_id = books.author_id
GROUP BY
    reviews.book_id,
    books.title,
    authors.author,
    subquery.avg_rating,
    subquery.n_ratings
ORDER BY n_reviews DESC
LIMIT 15;
'''

pd.io.sql.read_sql(query, con = engine)
```

<div class="alert alert-success">
<b>Reviewer's comment: </b> ‚úÖ
</div>

These are the top-15 books based on the number of text reviews they got and their average rating. Among them we see both recent bestsellers (Twilight, Harry Potter series, The Alchemist, Eat Pray Love etc.) and older books which by now can be considered classics of their genre (The Catcher in the Rye, The Hobbit or There and Back Again).

```python
# this function contains most of the query, and lets you display various parts of the dataset (beginning, end,
# ordered by different column etc.)

def reviews_rating_query(end_lines):
    query_text = '''
    SELECT
        DISTINCT reviews.book_id AS book_id,
        books.title,
        authors.author,
        COUNT(reviews.review_id) AS n_reviews,
        subquery.avg_rating,
        subquery.n_ratings
    FROM reviews
    LEFT JOIN
        (
        SELECT 
            DISTINCT book_id,
            AVG(rating) AS avg_rating,
            COUNT(DISTINCT rating_id) AS n_ratings
        FROM ratings 
        GROUP BY book_id
        ) AS subquery ON subquery.book_id = reviews.book_id
    LEFT JOIN books ON books.book_id = reviews.book_id
    LEFT JOIN authors ON authors.author_id = books.author_id
    GROUP BY
        reviews.book_id,
        books.title,
        authors.author,
        subquery.avg_rating,
        subquery.n_ratings
    '''
    full_query = query_text+end_lines
    return display(pd.io.sql.read_sql(full_query, con = engine))

# by the smallest number of reviews
print('Books with the smallest number of reveiws')
end_lines = '''
ORDER BY n_reviews
LIMIT 10;
'''
reviews_rating_query(end_lines)
print('')

# by the highest average rating
print('Books with the highest average rating')
end_lines = '''
ORDER BY avg_rating DESC
LIMIT 10;
'''
reviews_rating_query(end_lines)
print('')

# by the lowest average rating
print('Books with the lowest average rating')
end_lines = '''
ORDER BY avg_rating
LIMIT 10;
'''
reviews_rating_query(end_lines)
print('')
```

<a id='importanceofnumber'></a>

Interesting to notice that number of reviews and average rating don't seem to be strongly correlated. We see both books with high average rating but the small number of reviews (e.g. To Green Angel Tower by Tad Williams) as well as books with mediocre rating and high number of reviews (e.g. Twilight by Stephenie Meyer).

The key is, the less people rate the book, the more weight each individual rating has. **So when trying to assess the book by its average rating, we suggest you should also pay attention to the number of people who rated it.**

Perhaps, a good parameter to assess the book would be the ratio of reviews per ratings.

### <a id='activepublisher'>The Most Active Publisher</a>

Identify the publisher that has released the greatest number of books with more than 50 pages (this will help you exclude brochures and similar publications from your analysis).

```python
# This query shows only the first row. If you need the full table, please comment out (--) its last line (LIMIT 1;)

query = '''
SELECT
    books.publisher_id,
    publishers.publisher,
    COUNT(books.book_id) AS books_count
FROM books
LEFT JOIN publishers on publishers.publisher_id = books.publisher_id
WHERE num_pages > 50
GROUP BY
    books.publisher_id,
    publishers.publisher
ORDER BY books_count DESC
LIMIT 1;
'''

pd.io.sql.read_sql(query, con = engine)
```

### <a id='highestratedauthor'>The Highest Rated Author</a>

Identify the author with the highest average book rating: look only at books with at least 50 ratings.

```python
# This function contains most of the query, and lets you display various parts of the dataset (beginning, end,
# ordered by different column etc.)

def popular_author_query(end_lines):
    query_text = '''
    SELECT
        books.author_id,
        authors.author,
        COUNT(ratings.rating_id) AS n_ratings,
        AVG(ratings.rating) AS avg_rating
    FROM ratings
    LEFT JOIN books ON books.book_id = ratings.book_id
    LEFT JOIN authors ON authors.author_id = books.author_id
    GROUP BY
        books.author_id,
        authors.author
    '''
    full_query = query_text+end_lines
    return display(pd.io.sql.read_sql(full_query, con = engine))
    
end_lines='''
HAVING COUNT(ratings.rating_id) >= 50
ORDER BY avg_rating DESC
LIMIT 5;
'''
print('Ordered by average rating')
popular_author_query(end_lines)
print('')
    
end_lines='''
HAVING COUNT(ratings.rating_id) >= 50
ORDER BY
    n_ratings DESC,
    avg_rating DESC
LIMIT 5;
'''
print('Ordered by number of ratings and average rating')
popular_author_query(end_lines)
print('')

end_lines='''
HAVING COUNT(ratings.rating_id) >= 100
ORDER BY avg_rating DESC
LIMIT 5;
'''
print('Authors with 100 ratings or higher (ordered by average rating)')
popular_author_query(end_lines)
print('')
```

This is a direct consequence of [the fact we mentioned earlier](#importanceofnumber): the number of people who rated a book is extremely important.

If we calculate just the average book rating for every author, the highest rated writer is suddenly Diana Gabaldon, author of romance fantasy novels. However, this score doesn't reflect the real sutuation: it's based on 50 reviews only and means that Ms. Gabaldon has a small, but devoted fan club who rate all her books pretty high.

However, if we additionally order the table by the number of ratings (or simply set a higher entry level), we see the real winner of this book race, **J.K. Rowling**.

We'll achieve the same result also if we calculate average rating not per all books of the author, but per each single book.

```python
# This query shows only the first row. If you need the full table, please comment out (--) its last line (LIMIT 1;)

query = '''
SELECT
    books.author_id,
    books.book_id,
    books.title,
    authors.author,
    COUNT(ratings.rating_id) AS n_ratings,
    AVG(ratings.rating) AS avg_rating
FROM ratings
LEFT JOIN books ON books.book_id = ratings.book_id
LEFT JOIN authors ON authors.author_id = books.author_id
GROUP BY
    books.author_id,
    books.book_id,
    books.title,
    authors.author
HAVING COUNT(ratings.rating_id) >= 50
ORDER BY avg_rating DESC
LIMIT 1;
'''

pd.io.sql.read_sql(query, con = engine)
```

### <a id='avgreviewsnumber'>The Average Number of Text Reviews</a>

Find the average number of text reviews among users who rated more than 50 books.

```python
# average number

query = '''
SELECT AVG(n_reviews)
FROM
    (
    SELECT COUNT(DISTINCT review_id) AS n_reviews
    FROM reviews
    WHERE username IN
        (
        SELECT username
        FROM ratings
        GROUP BY username
        HAVING COUNT(DISTINCT book_id) > 50
        )
    GROUP BY username) AS subquery;
'''

pd.io.sql.read_sql(query, con = engine)
```

```python
# in details: how many reviews wrote each of these users

query = '''
SELECT
    username,
    COUNT(DISTINCT review_id) AS n_reviews
FROM reviews
WHERE username IN
        (
        SELECT username
        FROM ratings
        GROUP BY username
        HAVING COUNT(DISTINCT book_id) > 50
        )
GROUP BY username
ORDER BY n_reviews DESC;
'''

display(pd.io.sql.read_sql(query, con = engine))

query = '''
SELECT
    COUNT (DISTINCT username) AS n_reviewers
FROM reviews;'''

display(pd.io.sql.read_sql(query, con = engine))

query = '''
SELECT
    COUNT (DISTINCT username) AS n_users_who_rate_books
FROM ratings;'''

display(pd.io.sql.read_sql(query, con = engine))
```

## <a id='genconcl'>General Conclusions</a>

We've analysed the given datasets and came to certain conclusions which would be really improtant for those launching a book related app.

1) When selecting books for the app, you shouldn't rely solemnly on their rating. The rating itself is unrepresentative. You should always understand how many single grades it's based on.

2) You shouldn't expect app users to be very active. Only a small fraction of them generates content actively (leaves reviews, rates books). In this dataset, the share of active users was only 3.75%. If you expect user-generated content in your app, you should encourage it by offering some kind of promotions or discounts.

3) All the top-15 list was occupied by fiction books, both classics and recent bestsellers. We've noticed that our dataset contains several nonfiction books, but they have either low average rating or very little reviews. However, simply saying that fiction is more popular than nonfiction would be too far-fetched. It's possible that our dataset has a bigger share of fiction and lacks the biggest nonfiction names (Steven Hawking, Richard Dawkins, Dale Carnegie, Naomi Wolf etc.). With different selection of names and genres in the given dataset we could've come to a different conclusion.

4) And finally, some names you shouldn't miss out on: **Penguin Books** as one of the most popular and active book publishers, and **J.K. Rowling** as the highest rated author.
